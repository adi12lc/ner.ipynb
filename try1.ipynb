```python
from Decision_tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("diabetes.csv")
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Outcome'] = le.fit_transform(data['Outcome'])
X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1234)
```

```python
#Bootstraping
n_records = X_train.shape[0]
indexs = np.random.choice(n_records, n_records, replace=True) # randomly drawing samples from the records with replacement method
print(X_train[indexs], y_train[indexs])
```

```python
#fitting
n_estimators=10
trees = []
for i in range(n_estimators):
    dtree = DecisionTreeClassifier()
    X_sample, y_sample = X_train[indexs], y_train[indexs]
    dtree.fit(X_sample, y_sample)# fitting samples to the decision tree
    trees.append(dtree)#load tree to a list
```

```python
trees
```

#### Prediction

```python
predictions = np.swapaxes(np.array([dtree.predict(X_test) for dtree in trees]),0,1) # predict the test records for all the trees, swapaxes function transposes the array
```

```python
# function to find out most frequest values in an array 
def mode_label(x):
    vals,counts = np.unique(x, return_counts=True)
    index = np.argmax(counts)
    return vals[index]
```

```python
predictions = np.array([mode_label(i) for i in predictions]) # to convert prediction matrix to an array by majority voting method
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
